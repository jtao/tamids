```python
import pandas as pd 
import pandas_profiling

df = pd.read_csv("random_people.csv")
```

```python
#start getting a feel of the data
df.head(5)
```

```python
df['salary'].value_counts()
```

```python
df['salary'].median()
```

```python
def tax(s):
    if s>=6000:
        return s*.7
    else:
        return s*.85
```

```python
df["salary_after_tax"] = df["salary"].apply(tax)
df.head(5)
```

```python
df_high = df[df["salary_after_tax"]>10000]
```
